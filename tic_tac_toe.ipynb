# Tic Tac Toe Game (Without Any Library)

A clean, corrected implementation of the classic Tic Tac Toe game in Python.

**Fixes applied over the original code:**
- Renamed `sum()` to `check_sum()` to avoid shadowing Python's built-in `sum`
- Added **input validation** (non-numeric input, out-of-range, already-taken cell)
- Added **draw detection** when all 9 cells are filled with no winner
- Fixed `checkWin` return values so O winning is no longer falsy (`0`)
- Improved board separator alignment

```python
def check_sum(a, b, c):
    """Returns the sum of three values (used to check three-in-a-row)."""
    return a + b + c
```

```python
def print_board(x_state, o_state):
    """Prints the current state of the Tic Tac Toe board."""
    def cell(i):
        if x_state[i]:
            return 'X'
        elif o_state[i]:
            return 'O'
        else:
            return str(i)

    print(f" {cell(0)} | {cell(1)} | {cell(2)} ")
    print("---|---|---")
    print(f" {cell(3)} | {cell(4)} | {cell(5)} ")
    print("---|---|---")
    print(f" {cell(6)} | {cell(7)} | {cell(8)} ")
```

```python
def check_win(x_state, o_state):
    """
    Checks if there is a winner.
    Returns:
        'X'  -> X has won
        'O'  -> O has won
        None -> No winner yet
    """
    winning_combos = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # columns
        [0, 4, 8], [2, 4, 6]               # diagonals
    ]
    for combo in winning_combos:
        if check_sum(x_state[combo[0]], x_state[combo[1]], x_state[combo[2]]) == 3:
            return 'X'
        if check_sum(o_state[combo[0]], o_state[combo[1]], o_state[combo[2]]) == 3:
            return 'O'
    return None
```

```python
def is_draw(x_state, o_state):
    """Returns True if all cells are filled and there is no winner."""
    return all(x_state[i] or o_state[i] for i in range(9))
```

```python
def play_game():
    """Main function to run the Tic Tac Toe game."""
    x_state = [0] * 9
    o_state = [0] * 9
    turn = 1  # 1 for X, 0 for O

    print("Welcome to Tic Tac Toe!")
    print("Enter a number (0-8) to place your mark:")
    print(" 0 | 1 | 2 ")
    print("---|---|---")
    print(" 3 | 4 | 5 ")
    print("---|---|---")
    print(" 6 | 7 | 8 ")
    print()

    while True:
        print_board(x_state, o_state)
        current_player = 'X' if turn == 1 else 'O'
        print(f"\n{current_player}'s turn")

        # --- Input validation ---
        while True:
            raw = input("Please enter a position (0-8): ")
            if not raw.isdigit():
                print("Invalid input! Please enter a number between 0 and 8.")
                continue
            value = int(raw)
            if value < 0 or value > 8:
                print("Out of range! Please enter a number between 0 and 8.")
                continue
            if x_state[value] or o_state[value]:
                print("Cell already taken! Choose a different position.")
                continue
            break

        # --- Update state ---
        if turn == 1:
            x_state[value] = 1
        else:
            o_state[value] = 1

        # --- Check result ---
        winner = check_win(x_state, o_state)
        if winner:
            print_board(x_state, o_state)
            print(f"\n🎉 {winner} won the match! Congratulations!")
            break

        if is_draw(x_state, o_state):
            print_board(x_state, o_state)
            print("\nIt's a draw! Well played by both players.")
            break

        turn = 1 - turn

play_game()
```
